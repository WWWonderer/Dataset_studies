````python
answerable_contexts = []
answerable_questions = []
answerable_answers = []

unanswerable_contexts = []
unanswerable_questions = []
unanswerable_plausible_answers = []

for topic in squad_dev['data']:
    for passage in topic['paragraphs']:
        context = passage['context']
        for qa in passage['qas']:
            question = qa['question']
            if not qa['is_impossible']:
                for answer in qa['answers']:
                    answerable_contexts.append(context)
                    answerable_questions.append(question)
                    answerable_answers.append(answer['text'])
                    # take first answer only
                    break
            elif 'plausible_answers' in qa.keys():
                for plausible_answer in qa['plausible_answers']:
                    unanswerable_contexts.append(context)
                    unanswerable_questions.append(question)
                    unanswerable_plausible_answers.append(plausible_answer['text'])
                    # take first plausible answer only
                    break
````

````python
print("number of answerable questions: ", len(answerable_questions))
for i in range(5):
  print(answerable_contexts[i])
  print(answerable_questions[i])
  print(answerable_answers[i])
  print("*" * 1000)
````

````python
print("number of unanswerable questions: ", len(unanswerable_questions))
for i in range(5):
  print(unanswerable_contexts[i])
  print(unanswerable_questions[i])
  print(unanswerable_plausible_answers[i])
  print("*" * 1000)
````
