# Scifact dataset

```python
from beir import util, LoggingHandler
from beir.retrieval import models
from beir.datasets.data_loader import GenericDataLoader
from beir.retrieval.evaluation import EvaluateRetrieval
from beir.retrieval.search.dense import DenseRetrievalExactSearch as DRES

import logging
import pathlib, os
logging.basicConfig(format='%(asctime)s - %(message)s', datefmt='%Y-%m-%d %H:%M:%S', level=logging.INFO, force=True)

dataset = 'scifact'
url = "https://public.ukp.informatik.tu-darmstadt.de/thakur/BEIR/datasets/{}.zip".format(dataset)
out_dir = os.path.join(PROJECT_DIR, "datasets")
data_path = util.download_and_unzip(url, out_dir)

corpus, queries, qrels = GenericDataLoader(data_folder=data_path).load(split="test")

model = DRES(models.SentenceBERT("msmarco-distilbert-base-tas-b"), batch_size=16)
retriever = EvaluateRetrieval(model, k_values = [5], score_function="dot")
```

# examples of corpus, queries and qrels

```python
NUM_EXAMPLES = 3

def show_example(dict, num_examples):
  count = 0
  for key, value in dict.items():
    print(key, value)
    count += 1
    if count == num_examples: break

print("corpus length: ", len(corpus))
show_example(corpus, NUM_EXAMPLES)
print("-" * 1000)
print("queries length: ", len(queries))
show_example(queries, NUM_EXAMPLES)
print("-" * 1000)
print("qrels length: ", len(qrels))
show_example(qrels, NUM_EXAMPLES)
```

# examples of results

```python
results = retriever.retrieve(corpus, queries)
```

```python
show_example(results, NUM_EXAMPLES)
```

# evaluate results

```python
def eval_results(qrels, results, k_values):
  ndcg, _map, recall, precision = retriever.evaluate(qrels, results, k_values)
  print("ndcg: ", ndcg)
  print("map: ", _map)
  print("recall: ", recall)
  print("precision: ", precision)

eval_results(qrels, results, retriever.k_values)
```

# other retriever methods (DPR)



```python
from transformers import AutoTokenizer

model = AutoTokenizer.from_pretrained('facebook/dpr-question_encoder-single-nq-base')
print(type(model))

model = AutoTokenizer.from_pretrained('facebook/dpr-ctx_encoder-single-nq-base')
print(type(model))
```

```python
model = DRES(models.DPR(("facebook/dpr-question_encoder-single-nq-base", "facebook/dpr-ctx_encoder-single-nq-base")), batch_size=16)

retriever = EvaluateRetrieval(model, k_values=[5, 10], score_function="dot")
results = retriever.retrieve(corpus, queries)
eval_results(qrels, results, retriever.k_values)
```

```python
from beir.retrieval.search.lexical import BM25Search as BM25
from beir.retrieval.evaluation import EvaluateRetrieval

#### Provide parameters for elastic-search
hostname = "localhost"
index_name = "scifact"
initialize = True # True, will delete existing index with same name and reindex all documents

model = BM25(index_name=index_name, hostname=hostname, initialize=initialize)
retriever = EvaluateRetrieval(model)

#### Retrieve dense results (format of results is identical to qrels)
results = retriever.retrieve(corpus, queries)
```
